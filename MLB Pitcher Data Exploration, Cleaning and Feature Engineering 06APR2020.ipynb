## PITCHES.CSV

```python
#load ALL MLB pitch data for 2015-2018 using pitches.csv
pitches = pd.read_csv('pitches.csv')
pitches.describe()
```

```python
#drop all NaNs
pitches.dropna(inplace=True)
```

### Pitch types thrown by all pitchers (by percentage)

```python
#look at all pitch types thrown by all pitchers
allptypes = pitches['pitch_type']
allptypes.value_counts() / len(allptypes) * 100
```

```python
#drop rows of all pitches that are pitchouts/and intentional balls and unknown pitch types
pitchesrmv = ['FO', 'PO', 'IN', 'UN', 'FA', 'AB']

for i in pitchesrmv:
    pitches = pitches[pitches['pitch_type']!=i]

#look at the counts of all pitch types for all pitchers    
pitches['pitch_type'].value_counts()
```

## Remap all pitch types to 3 categories of pitches

```python
#dictionary for pitch types to reduce pitches for ALL pitches thrown by all pichers
#FB (fastball type pitches) = FF (four-seam fastball), FT (two-seam fastball), FC (cut fastball), SI (sinker), 
#BR (breaking ball type pitches) = SL (slider), CU (curveball), SC (screwball)
#OS (offspeeed type pitches) = CH (change-up), FS (split-finger fastball), KB (knuckleball), EP (eephus)

globalpitchmap = {'FF':'FB', 'SL':'BR', 'FT':'FB', 'CH':'OS', 'SI':'OS', 
                  'CU':'BR', 'FC':'FB', 'KC':'BR', 'FS':'OS', 'KN':'OS', 'EP':'OS', 'SC':'BR'}

pitches = pitches.applymap(lambda s: globalpitchmap.get(s) if s in globalpitchmap else s)
pitches['pitch_type'].value_counts()
```

```python
#show as a percentage
pitches['pitch_type'].value_counts() * 100 / len(pitches['pitch_type'])
```

### ALL Pitches belonging to Pitcher ID

```python
#select ALL pitches from the list of AT BATS in atbat_id - these are the pitches thrown by the pitcher
mypitches = pitches[pitches['ab_id'].isin(atbat_id)]
mypitches.describe()
```

```python
#use this to avoid later problems with having * in the string
mypitches = mypitches.replace('*B','DirtB')
```

###  Drop UNNECESSARY Columns

```python
#drop unnecessary columns
dropcols = ['start_speed', 'spin_rate', 'spin_dir', 'break_angle', 'break_length', 'break_y',
           'ax', 'ay', 'az', 'sz_bot', 'sz_top', 'type_confidence', 'vx0', 'vy0', 'vz0', 'x', 'x0', 'y',
           'y0', 'z0', 'pfx_x', 'pfx_z', 'nasty', 'zone', 'event_num','code']

mypitches = mypitches.drop(columns=dropcols, axis=1)
```

### Pitch Type Frequency for Pitcher

```python
#Frequency of pitches thrown by the selected pitcher
mypitchtypes = mypitches['pitch_type']
order = mypitchtypes.value_counts() / len(mypitchtypes) * 100
order
```

## LABEL ENCODING

```python
#establish a dictionary for the 3 categories of pitches
pitchmap = {'FB': 1, 'BR': 2, 'OS': 3}
```

```python
#replace the pitch_type column 2 letter codes with a correspending value from the dictionary pitchmap,
#then add in the old 2 letter column as a guide for later look up

pitchtypetmp = mypitches.copy()['pitch_type']
mypitches = mypitches.applymap(lambda s: pitchmap.get(s) if s in pitchmap else s)
mypitches['pitch_type_code'] = pitchtypetmp
mypitches['pitch_type'] = mypitches['pitch_type'].astype(int)
mypitches.shape
```

### TYPE CONVERSIONS

```python
#convert ab_id to integer
mypitches['ab_id'] = mypitches['ab_id'].astype(int)
```

```python
#add a years column to mypitches
mypitches = addYear(mypitches)
```

```python
#Duplicates values from a unique ab_id from a specific at bat (column) from the atbat dataframe
#to another dataframe with multiple instances of ab_id values (multiple values in the same column).

def copyData(from_df, to_df, atbatidlist, column_name):
    newcol = column_name
    to_df[newcol] = 0
    tmp = from_df[['ab_id', column_name]]
    
    for i in range(len(atbatidlist)):
        atbatid = np.array(atbatidlist)[i]
        val = tmp.loc[tmp['ab_id']==atbatid, newcol]
        val = np.array(val)[0]
        to_df.loc[to_df['ab_id']==atbatid, newcol] = val
        
    return to_df
```

```python
#copy data
mypitches = copyData(atbats_pitcher, mypitches, atbat_id, 'stand')
mypitches = copyData(atbats_pitcher, mypitches, atbat_id, 'inning')
mypitches = copyData(atbats_pitcher, mypitches, atbat_id, 'p_score')
```

## ONE HOT ENCODING

```python
#one hot encode columns (pitch result ('type') and batter handedness ('stand'))
mypitches = pd.get_dummies(mypitches, columns=['type','stand'])
mypitches
```

# WRITE OUT DATA

```python
output = mypitches
output.to_csv('PitcherData.csv', index=False)
```
